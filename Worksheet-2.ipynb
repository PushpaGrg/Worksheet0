Problem 2 - Data Imputations:
Complete all the following Task:
• Dataset for the Task: "medical_student.csv"
1. Load the provided dataset and import in pandas DataFrame.
2. Check info of the DataFrame and identify column with missing (null) values.
3. For the column with missing values fill the values using various techniques we discussed above. Try
to explain why did you select the particular methods for particular column.
4. Check for any duplicate values present in Dataset and do necessary to manage the duplicate items.
{Hint: dataset.duplicated.sum()}

```python
import pandas as pd
#Read csv file into dataframe df2 from specified file path
df2=pd.read_csv('/content/drive/MyDrive/Datasets/medical_students_dataset.csv')
df2
```

```python
#Display general info of the dataframe
df2.info()
```

```python
#Check for total number of missing values
df2.isnull().sum()
```

```python
#Forward fill for categorical columns
df2['Gender']=df2['Gender'].ffill()
df2['Blood Type']=df2['Blood Type'].ffill()
df2['Diabetes']=df2['Diabetes'].ffill()
df2['Smoking']=df2['Smoking'].ffill()
#Dill missing values in numeric columns with mean
df2['Age']=df2['Age'].fillna(df2['Age'].mean())
df2['Weight']=df2['Weight'].fillna(df2['Weight'].mean())
df2['Temperature']=df2['Temperature'].fillna(df2['Temperature'].mean())
df2['Heart Rate']=df2['Heart Rate'].fillna(df2['Heart Rate'].mean())
df2['Blood Pressure']=df2['Blood Pressure'].fillna(df2['Blood Pressure'].mean())
df2['Cholesterol']=df2['Cholesterol'].fillna(df2['Cholesterol'].mean())
#Fill missing values in 'Height' and 'BMI' columns with median
df2['Height']=df2['Height'].fillna(df2['Height'].median())
df2['BMI']=df2['BMI'].fillna(df2['BMI'].median())
#Fill missing values in 'Temperature' with 0(if appropriate)
df2['Temperature']=df2['Temperature'].fillna(0)
```

-->For categorical columns, forward fill is used because it assumes missing values can be filled with last valid entry,especially in sequential data or forms with gaps.

--> For numeric columns, mean is used because it assumes that the data is roughly normal or symmetric without significant outliers.

--> For Height and BMI, median is used because it is less affected by extreme values since these columns may have skewed distributions or outliers.
-->If missing values in temperature represent absence of data then those values replaced with 0 is reasonable.

```python
#Check that there are no missing values anymore
df2.isnull().sum()
```

Problem - 1:
Create a DataFrame that is subsetted for the columns ’Name’, ’Pclass’, ’Sex’, ’Age’, ’Fare’, and ’Survived’.
Retain only those rows where ’Pclass’ is equal to 1, representing first-class passengers. What is the mean,
median, maximum value, and minimum value of the ’Fare’ column?

```python
import pandas as pd
#Load 'titanic.csv' into dataframe
df3=pd.read_csv('/content/drive/MyDrive/Datasets/Titanic-Dataset.csv')
#Create a subset of dataframe containing only relevant columns
df3_subset=df3[['Name','Pclass','Sex','Age','Fare','Survived']]
#Filter dataframe to keep only first-class passengers
first_class=df3_subset[df3_subset['Pclass']==1]
#Compute mean of 'Fare' column for first-class passengers
mean=first_class['Fare'].mean()
print("Mean Fare: ",mean)
#COmpute median
median=first_class['Fare'].median()
print("Median Fare: ",median)
#Compute maximum value
maximum=first_class['Fare'].max()
print("Maximum Fare: ",maximum)
#Compute minimum value
minimum=first_class['Fare'].min()
print("Minimum Fare: ",minimum)
```

Problem - 3:

The ’Embarked’ column in the Titanic dataset contains categorical data representing the ports of embarka-
tion:

• ’C’ for Cherbourg
• ’Q’ for Queenstown
• ’S’ for Southampton
Task:
1. Use one-hot encoding to convert the ’Embarked’ column into separate binary columns (’Embarked C’,
’Embarked Q’, ’Embarked S’).
2. Add these new columns to the original DataFrame.
3. Drop the original ’Embarked’ column.

```python
#Use one-hot encoding to 'Embarked' column
df3_encoding=pd.get_dummies(df3['Embarked'],prefix='Embarked')
#Concatenate new encoded columns with orginal dataframe
df3=pd.concat([df3, df3_encoding],axis=1)
#Drop original 'Embarked' column
df3=df3.drop('Embarked', axis=1)
#Print the first rows of modified dataframe
print(df3.head())
```

Compare the mean survival rates (’Survived’) for the different groups in the ’Sex’ column. Draw a visual-

```python
import matplotlib.pyplot as plt
import seaborn as sns
#Group the dataframe by 'Sex' and calculate mean of 'Survived' column for each group
survival_mean=df3.groupby('Sex')['Survived'].mean()
#Display the survival rates for each gender
survival_mean
```

Problem - 6{Optional}:

Show how the survival rates (’Survived’) vary by age group and passenger class (’Pclass’). Break up the
’Age’ column into five quantiles in your DataFrame, and then compare the means of ’Survived’ by class
and age group. Draw a visualization using a any plotting library to represent this graphically.

```python
#Create 5 equal-sized quantile-based age groups
df3['AgeGroup']=pd.qcut(df3['Age'],5,labels=['Q1','Q2','Q3','Q4','Q5'])
#Calculate mean survival rate for each combination of AgeGroup and Plass
survival_rate=df3.groupby(['AgeGroup','Pclass'])['Survived'].mean().reset_index()
#Create a figure of 10X6 inches
plt.figure(figsize=(10,6))
#Create a bar plot to show mean survival rates by AgeGroup and Pclass
sns.barplot(x='AgeGroup',y='Survived',hue='Pclass',data=survival_rate,palette='muted')
#Set title of the plot
plt.title('Mean Survival Rates by Age Group and Passenger Class',fontsize=17)
plt.xlabel('Age Group',fontsize=13)  #label x-axis
plt.ylabel('Mean Survival Rate',fontsize=13)  #label y-axis
#Customize legend to 'Passenger Class and place it in the upper left corner'
plt.legend(title='Passenger Class',loc='upper left')
plt.show() #Display the plot
```
